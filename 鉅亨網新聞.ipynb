```python
from google.colab import drive
import os

drive.mount('/content/drive/', force_remount=True)
os.chdir('/content/drive/My Drive/Data_Technology/DT_Learning/III_Big_Data/BDSE21_Team1/Stock_Analyzer_Model_2021_01/')

def news_crawler_anue(year):
    date_start, date_stop = str(year) + '-01-01', str(year) + '-12-31'

    from datetime import datetime, timedelta
    date_stop = str(datetime.strptime(date_stop, '%Y-%m-%d') - timedelta(days=1))[:10]

    # Get data from anue
    from anuecrawler.news import headline, twstock, wdstock, blockchain
    import pandas as pd
    import numpy as np
    hd = twstock
    hd = hd.browse(date_start, date_stop)
    hd = hd.query(['newsId', 'market', 'title', 'summary', 'content'])
    df = pd.DataFrame(hd)

    # Get datetime column with multi-threads
    df['datetime'] = np.empty(len(df))

    from concurrent.futures import ThreadPoolExecutor, as_completed
    with ThreadPoolExecutor(max_workers=30) as executor:
        future = {executor.submit(get_data_datetime, df, i): i for i in range(len(df))}
    
    df = df.reindex(columns=['datetime', 'newsId','market', 'title', 'summary', 'content'])

    # Arrange the dataframe
    df = df.drop(df[df.market.map(len) == 0].index)

    cancel_text = 'a-zA-Z0-9\/\.\-\=\_\;\:\&\%\?\ \(\)'
    df['content'] = pd.Series(df['content']).str.replace(f'&[{cancel_text}]+t;', '', case=False, regex=True)
    df['content'] = pd.Series(df['content']).str.replace(f'&[{cancel_text}]+p;', '', case=False, regex=True)
    df['content'] = pd.Series(df['content']).str.replace(f'&[{cancel_text}]+h;', '', case=False, regex=True)
    df['content'] = pd.Series(df['content']).str.replace(f'http[{cancel_text}]+', '', case=False, regex=True)
    df['content'] = pd.Series(df['content']).str.replace(f'a href=[{cancel_text}]+', '', case=False, regex=True)
    df['content'] = pd.Series(df['content']).str.replace(f'img [{cancel_text}]+', '', case=False, regex=True)
    df['content'] = pd.Series(df['content']).str.replace(f'data-height[{cancel_text}]+', '', case=False, regex=True)

    df = df.dropna()
    df = df.sort_values('datetime')
    df['datetime'] = df['datetime'].astype(int)
    df = df.drop(df[df.datetime == 0].index)
    df = df.reset_index(drop=True)

    # Save to csv
    filename = f'Data_News/News_Anue_{year}.csv'
    df.to_csv(filename)

    del df

def get_data_datetime(df, i):
    import requests
    from bs4 import BeautifulSoup
    url = 'https://news.cnyes.com/news/id/'

    response = requests.get(url + str(df['newsId'][i]))
    soup = BeautifulSoup(response.text, 'lxml')
    response.close()
    try:
        datetime = soup.find('time', attrs={'data-reactid': '234'}).text[:10].replace('/', '')
    except Exception as e:
        datetime = soup.find('time', attrs={'data-reactid': '230'}).text[:10].replace('/', '')
    df['datetime'][i] = int(datetime)

def get_data_stock(stock_id, year):
    date_start, date_stop = str(year) + '-01-01', str(year) + '-12-31'

    from FinMind.data import DataLoader
    dl = DataLoader()
    stock_data = dl.taiwan_stock_daily(
        stock_id=stock_id, start_date=start_date, end_date=end_date
    )
    stock_data = dl.feature.add_kline_institutional_investors(
        stock_data
    )
    stock_data = dl.feature.add_kline_margin_purchase_short_sale(
        stock_data
    )

    # Save to csv
    filename = f'Data_Stock/{stock_id}_{year}.csv'
    stock_data.to_csv(filename)
    del stock_data

def arrange_by_datetime_and_stock_code(year):
    import pandas as pd
    import json
    filename = f'Data_News/News_Anue_{year}.csv'
    df_read = pd.read_csv(filename, index_col=0)

    datetime_list = df_read['datetime']
    code_list = df_read['market']

    df = []

    for i in range(len(datetime_list)):
        code_list[i] = code_list[i].replace('\'', '"')
        code_list[i] = json.loads(code_list[i])
        for market_info in code_list[i]:
            df.append([datetime_list[i], market_info['code']])
    
    df = pd.DataFrame(df)
    df = df.sort_values([0, 1])
    df = df.drop_duplicates()
    df = df.reset_index(drop=True)
    df[2] = None
    df.rename(columns={0: 'datetime', 1: 'stock_code', 2: 'content'}, inplace=True)

    # Insert the news data with multi-threads
    from concurrent.futures import ThreadPoolExecutor, as_completed
    with ThreadPoolExecutor(max_workers=30) as executor:
        future = {executor.submit(insert_news_data_to_new_dataframe, df, i): i for i in range(len(datetime_list))}

    return df

def insert_news_data_to_new_dataframe(df, i):
    code_list[i] = code_list[i].replace('\'', '"')
    code_list[i] = json.loads(code_list[i])
    for market_info in code_list[i]:
        df['content'][df[df[(df.datetime == i) & (df.stock_code == market_info['code'])].index]] += df_read['summary'][i]

if __name__ == '__main__':
    year = 2021
    # news_crawler_anue(year)
    df = arrange_by_datetime_and_stock_code(year)
    # get_data_stock(2330, year)
```
