```python
from __future__ import division, unicode_literals 
import codecs
from bs4 import BeautifulSoup

f=codecs.open("tsmc_url.html", "r", encoding='utf-8')
document= BeautifulSoup(f.read())
print(document)
```

```python
url_list=[]
for link in document.find_all('a'):
    url_list.append(link.get('href'))
    print(link.get('href'))
```

```python
url_list
```

```python
data={'date':[],'issued_share':[]}
df_2330=pd.DataFrame(data)
```

```python
def insert_df(df):
  df=df.drop([0,1,2,3])
  df=df.rename(columns={'股票名稱':'code', '◎發行股份':'issued_share'})
  share=df['issued_share'][df.code=='2330  '].values[0]
  return share
```

```python
for item in url_list:
  resp = urlopen('https://www.twse.com.tw/'+item)
  zipfile = ZipFile(BytesIO(resp.read()))
  zipfile.namelist()
  date=zipfile.namelist()[0][3:9]
  df = pd.read_excel(zipfile.open(zipfile.namelist()[0]).read(),sheet_name='7',header=1,usecols='A,L')  # <-- add .read()
  share=insert_df(df)
  df_2330.append({'date': date,'issued_share': share}, ignore_index=True)
```
